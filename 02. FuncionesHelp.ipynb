Los parametros opcionales siempren deben de ir al final

```python
def greet(who="Colin"):
    print ("Hello, " , who)

greet()
greet(who="Kaggle")
greet(who = "World")
```

```python
def multiplicar_por_cinco(x):
    return 5 * x

def call (fn, arg):
    "'llama a la funcion con argumento'"
    return fn(arg)

def cuadrado(fn, arg):
    "'Call fn on the result of calling fn on arg'"
    return fn(fn(arg))

print(
    call(multiplicar_por_cinco, 1),
    cuadrado(multiplicar_por_cinco, 1),
    sep = '\n',
)
```

Estas funcionnes que toman como parametros otras funciones se les denomia funciones de orden superior
